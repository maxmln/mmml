```python
# Import necessary packages
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import sklearn
from sklearn.linear_model import Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Conv1D, LSTM, Dense
from keras.callbacks import LearningRateScheduler
from tensorflow.losses import huber_loss
from keras.optimizers import SGD
```

```python
# Import APPL historical data
import yfinance as yf
aapl = yf.Ticker("aapl")
df = aapl.history(period="max")
```

```python
df.head()
```

```python
df['Close'].plot(label='AAPL', figsize=(16,8), title='Adjusted Closing Price', grid=True)
```

```python
window_size = 32 

num_samples = len(df) - window_size
indices = np.arange(num_samples).astype(np.int)[:,None] + np.arange(window_size + 1).astype(np.int)
```

```python
data = df['Close'].values[indices] 
```

```python
X = data[:,:-1]
y = data[:,-1] 
```

```python
# Train and test split
split_fraction = 0.75
ind_split = int(split_fraction * num_samples)
X_train = X[:ind_split]
y_train = y[:ind_split]
X_test = X[ind_split:]
y_test = y[ind_split:]
```

# Ridge Regression

```python
ridge_model = Ridge()
ridge_model.fit(X_train, y_train)

y_pred_train_ridge = ridge_model.predict(X_train)
y_pred_ridge = ridge_model.predict(X_test)
```

```python
# Training data
df_ridge = df.copy()
df_ridge.drop(['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits',], axis=1, inplace=True)
df_ridge = df_ridge.iloc[window_size:ind_split]
df_ridge['Close Train'] = y_pred_train_ridge[:-window_size]
df_ridge.plot(label='AAPL', figsize=(16,8), title='Closing', grid=True)
```

```python
# Test data
df_ridge = df.copy()
df_ridge.drop(['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits',], axis=1, inplace=True)
df_ridge = df_ridge.iloc[window_size+ind_split:] # Past 32 days we don't know yet
df_ridge['Close Train'] = y_pred_ridge
df_ridge.plot(label='AAPL', figsize=(16,8), title='Closing', grid=True)
```

# Gradient Boosting Trees

```python
gb_model = GradientBoostingRegressor()
gb_model.fit(X_train, y_train)

y_pred_train_gb = gb_model.predict(X_train)
y_pred_gb = gb_model.predict(X_test)
```

```python
#Training data
df_gb = df.copy()
df_gb.drop(['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits',], axis=1, inplace=True)
df_gb = df_gb.iloc[window_size:ind_split]
df_gb['Close Train'] = y_pred_train_gb[:-window_size]
df_gb.plot(label='AAPL', figsize=(16,8), title='Closing Price', grid=True)
```

```python
# Test data
df_gb = df.copy()
df_gb.drop(['Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits'], axis=1, inplace=True)
df_gb = df_gb.iloc[ind_split+window_size:]
df_gb['Close Test'] = y_pred_gb
df_gb.plot(label='AAPL', figsize=(16,8), title='Closing Price', grid=True)
```

# Gradient Boosting Trees

```python
keras_model = Sequential([
  Conv1D(filters=128, kernel_size=5, input_shape=(window_size, 1), strides=1, padding="causal", activation="tanh"),
  Conv1D(filters=128, kernel_size=5, strides=1, padding="causal", activation="tanh"),
  LSTM(128, return_sequences=True),
  LSTM(128),
  Dense(64, activation="tanh"),
  Dense(64, activation="tanh"),
  Dense(1)
])

keras_model.summary()
```

```python
# First figure out the right learning rate
lr_schedule = LearningRateScheduler(lambda epoch: 1e-8 * 10**(epoch / 20))
optimizer = SGD(lr=1e-8, momentum=0.9)
keras_model.compile(loss=huber_loss,
                    optimizer=optimizer,
                    metrics=["mae"])
# Scale the data due to LSTM units
scaler = MinMaxScaler(feature_range=(-1, 1))
data_transform = scaler.fit_transform(df['Close'].values[:,None])
data_transform = data_transform[indices]
X2 = data_transform[:,:-1]
y2 = data_transform[:,-1]
X_train_reshape = X2[:ind_split]
y_train_reshape = y2[:ind_split]
X_test_reshape = X2[ind_split:]
y_test_reshape = y2[ind_split:]
```

```python
history = keras_model.fit(X_train_reshape,
    y_train_reshape, 
    batch_size=256, epochs=250, 
    callbacks=[lr_schedule], 
    validation_data=(X_test_reshape, y_test_reshape)
)
```

```python
plt.semilogx(history.history['lr'], history.history['val_loss'])
plt.legend(['Learning Rate', 'Validation Loss'])
```
